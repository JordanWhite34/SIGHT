```python
config = {
    "path": "/working/dataset",
    "train": "/working/dataset/train/images",
    "val": "/working/dataset/val/images",
    "predict": "/working/dataset/test/images",
    "nc": 1,
    "names": ["human"]
}

with open("config.yaml", "w") as f:
    yaml.dump(config, f)

with open("config.yaml", "r") as f:
    print(f.read())
```

```python
def main():
    model = YOLO("yolov8n.pt")
    model.train(
        # Project
        project="Polar-Owl",
        name="yolov8n",
        deterministic=True,
        seed=42,
        data="working/config.yaml",
        save=True,
        save_period=5,
        pretrained=True,
        imgsz=1280,
        epochs=20,  # Increased from 10 to 20
        batch=8,    # Increased batch size
        workers=16,  # Adjusted workers
        val=True,
        device=torch.device('mps'),
        lr0=0.01,  # Adjusted learning rate
        patience=3,
        optimizer="Adam",  # Changed optimizer to Adam
        momentum=0.9,  # Adjusted for Adam optimizer
        weight_decay=0.001,  # Increased weight decay
        close_mosaic=5,
    )
```

```python
if __name__ == '__main__':
    main()
```

```python
import cv2
import matplotlib.pyplot as plt
import torch

# Assuming 'model' is already loaded and configured correctly
# If the model expects a different format, ensure to adapt the preprocessing accordingly

# Load the video
video_path = "working/dataset/realworld/video/2.MOV"
cap = cv2.VideoCapture(video_path)

# Check if video opened successfully
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Loop through the frames of the video
while True:
    ret, frame = cap.read()
    if not ret:
        break  # Break the loop if there are no frames left to read

    # Convert frame to the format expected by your model if needed, typically RGB
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Use your model's prediction method
    results = model.predict(frame_rgb, conf=0.5)[0]

    # Plotting the results
    image = results.plot()

    # Convert back to BGR for OpenCV
    image_bgr = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

    # Display the frame
    cv2.imshow('Video', image_bgr)

    # Exit if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
